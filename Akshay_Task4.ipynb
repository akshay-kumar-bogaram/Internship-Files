```python
## this is a function to replace values which are less than threshold to others as we cannot plot the single count values

threshold = 50
def replace(df, column_name):
    values = df[column_name].value_counts()
    replacement = "Other"
    cols_replace = values[values < threshold].index

    df[column_name] = df[column_name].apply(lambda x: replacement if x in cols_replace else x)
    return df[column_name]
```

### Average salary according to age

```python
def split(df, column_name):
    empty = []
    for i in df[column_name]:
        empty.append(i.split()[0])
    df[column_name] = empty
```
